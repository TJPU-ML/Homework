```python
def gradicent_train( X ,Y,epoch=1000, test_size = 0.3,random_state= 1111):  #定义梯度下降函数，
    
    X_train,X_test,Y_train,Y_test = train_test_split(
        X ,Y, test_size=test_size, random_state= random_state)#对iris数据集进行数据分割，分出数据集和测试集

    
    w = np.ones((X_train.shape[1],Y_train.shape[1]))   
    error_train = np.ones((epoch,))
    error_test = np.ones((epoch,))

    
    #训练数据模型
    for i in range(epoch): #创建整数列表（epoch）  迭代函数
        error_train[i],error_test[i] = pre_error(
            X_train,X_test,Y_train,Y_test,w)   #记录数据
        error = pre(X_train,w) - Y_train  
        w = w - X_train.transpose() * error      #权值的更新，为了使同一层中的weight和bias会以不同的学习率进行更新
    return w,error_train,error_test
```
